```python
from nsepy import get_history
from tqdm.notebook import tqdm
import os
import pandas as pd
from datetime import datetime
```

```python
listed_companies = pd.read_excel('templates/MCAP31032021_0.xlsx')
listed_companies = listed_companies.rename(columns={'Market capitalization \n(Rs in Lakhs)': 'Cap'})
cols_interested = ['Symbol', 'Cap']
listed_companies_fil = listed_companies[cols_interested].head(1500)
listed_companies_fil['Cap'] = listed_companies_fil['Cap'].astype(float)
listed_companies_fil = listed_companies_fil[listed_companies_fil['Cap'] >= 75000]
interested_symbols = listed_companies_fil[cols_interested[0]].values.tolist()
```

```python
end_date = datetime.today()
start_date = end_date.replace(year=end_date.year-7)
```

```python
for symbol in tqdm(interested_symbols):
    output_file = f'market_data/{symbol}.csv'
    if not os.path.exists(output_file):
        get_history(symbol, start=start_date, end=end_date).to_csv(output_file)
    else:
        continue
```

```python
sector_data = [line.strip().split("=") for line in open('sector_data.txt').readlines()]
companies_list = []
for sector_info in sector_data:
    sector, companies = sector_info
    for company in companies.split(","):
        companies_list.append((company.strip(), sector))
df = pd.DataFrame(companies_list, columns=['Company Name', 'Sector'])
df.to_csv('sector_data.csv', index=False)
```

```python
df = pd.read_csv('sector_data.csv')
listed_companies['Company Name'] = listed_companies['Company Name'].astype(str)
df['Company Name'] = df['Company Name'].astype(str)
companies_list = []
for index, row in df.iterrows():
    name, sector = row
    filtered = listed_companies[listed_companies['Company Name'].str.lower().str.contains(name.lower())]
    if len(filtered) > 0:
        companies_list.append((filtered.iloc[0]['Company Name'], sector))
    else:
        continue
df = pd.DataFrame(companies_list, columns=['Company Name', 'Sector'])
listed_companies = listed_companies.merge(df, on='Company Name')[['Symbol', 'Company Name', 'Cap', 'Sector_y']]
listed_companies.columns = ['Symbol', 'Company Name', 'Cap', 'Sector']
```
